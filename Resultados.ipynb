**Criação dos Data-Frames dos grupos de doenças mentais**

```r
bipolar_data1 <- data.frame("x1" = dados$BD1,
                            "x2" = dados$...3,
                            "x3" = dados$...4,
                            "x4" =  dados$...5,
                            "x5" = dados$...6,
                            "x6" = dados$...7,
                            "x7" = dados$...8)

row.names(bipolar_data1) <- dados$Label
```

```r
bipolar_data1
```

```r
bipolar_data2 <- data.frame("x1" = dados$BD2,
                            "x2" = dados$...10,
                            "x3" = dados$...11,
                            "x4" =  dados$...12,
                            "x5" = dados$...13,
                            "x6" = dados$...14,
                            "x7" = dados$...15,
                            "x8" = dados$...16)

row.names(bipolar_data2) <- dados$Label
```

```r
bipolar_data2
```

```r
controle_data <- data.frame("x1" = dados$HC,
                            "x2" = dados$...18,
                            "x3" = dados$...19,
                            "x4" =  dados$...20,
                            "x5" = dados$...21,
                            "x6" = dados$...22,
                            "x7" = dados$...23,
                            "x8" = dados$...24,
                            "x9" = dados$...25,
                            "x10" = dados$...26,
                            "x11" = dados$...27)

row.names(controle_data) <- dados$Label
```

```r
controle_data
```

```r
esquizo_data <- data.frame("x1" = dados$SZC,
                            "x2" = dados$...29,
                            "x3" = dados$...30,
                            "x4" =  dados$...31,
                            "x5" = dados$...32,
                            "x6" = dados$...33,
                            "x7" = dados$...34,
                            "x8" = dados$...35,
                            "x9" = dados$...36,
                            "x10" = dados$...37,
                            "x11" = dados$...38)

row.names(esquizo_data) <- dados$Label
```

```r
esquizo_data
```

```r
for(i in c(1:length(dados$Label))){

    controle <- controle_data  %>% slice(i) %>% t() %>% as.data.frame() %>%  mutate(label = "Controle") 
    esquizo <- esquizo_data  %>% slice(i) %>% t() %>% as.data.frame() %>% mutate(label = "Esquizofrenia") 
    bipolar1 <- bipolar_data1  %>% slice(i) %>% t() %>% as.data.frame() %>% mutate(label = "Bipolar_1")
    bipolar2 <- bipolar_data2  %>% slice(i) %>% t() %>% as.data.frame() %>% mutate(label = "Bipolar_2")
    group <- rbind(controle, esquizo, bipolar1,bipolar2) %>% rename(value = V1)
    
    plot <-  ggplot(data = group, aes(x = label, y = value, color = as.factor(label))) + 
        geom_boxplot() + geom_jitter(aes(alpha = 0.5)) + theme_fivethirtyeight() + 
        labs(title= paste("Protein: ",dados$Label[i])) +
        xlab("Groups") +
        ylab("Log2 (Intensity)") +
        scale_fill_discrete(name = "EXPLANATION") +
        theme(legend.position = "none")             
        
        print(plot)
}

```
