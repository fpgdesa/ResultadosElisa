**Criação dos Data-Frames dos grupos de doenças mentais**

```r
bipolar_data <- data.frame("x1" = dados$BD,
                            "x2" = dados$...3,
                            "x3" = dados$...4,
                            "x4" =  dados$...5,
                            "x5" = dados$...6,
                            "x6" = dados$...7,
                            "x7" = dados$...8)

row.names(bipolar_data) <- dados$Grupos
```

```r
bipolar_data
```

```r
controle_data <- data.frame("x1" = dados$HC,
                            "x2" = dados$...10,
                            "x3" = dados$...11,
                            "x4" =  dados$...12,
                            "x5" = dados$...13,
                            "x6" = dados$...14,
                            "x7" = dados$...15,
                            "x8" = dados$...16,
                            "x9" = dados$...17,
                            "x10" = dados$...18,
                            "x11" = dados$...19,
                            "x12" = dados$...20,
                            "x13" = dados$...21)

row.names(controle_data) <- dados$Grupos
```

```r
controle_data
```

```r
esquizo_data <- data.frame("x1" = dados$SCZ,
                            "x2" = dados$...23,
                            "x3" = dados$...24,
                            "x4" =  dados$...25,
                            "x5" = dados$...26,
                            "x6" = dados$...27,
                            "x7" = dados$...28)

row.names(esquizo_data) <- dados$Grupos
```

```r
esquizo_data
```

```r
for(i in c(1:length(dados$Grupos))){

    controle <- controle_data  %>% slice(i) %>% t() %>% as.data.frame() %>%  mutate(label = "HC") 
    esquizo <- esquizo_data  %>% slice(i) %>% t() %>% as.data.frame() %>% mutate(label = "ED") 
    bipolar <- bipolar_data  %>% slice(i) %>% t() %>% as.data.frame() %>% mutate(label = "BD")
    group <- rbind(controle, esquizo, bipolar) %>% rename(value = V1)
    
    plot <-  ggplot(data = group, 
                        aes(x = label, 
                        y = value,
                        #alpha = 0.2, 
                        colour = as.factor(label))) + 
        
        geom_boxplot(aes(fill = as.factor(label)),
                     outlier.colour = "white",
                     alpha = 0.4,
                     colour = "black") + 
        
        geom_jitter() + 
        
        theme(legend.position = "none")+
        
        labs(x= "Diagnosis",
             y= paste("Protein ",
                      dados$Grupos[i], 
                      "(log2)"), 
            title= NULL) +
        
        theme(
        axis.line = element_line(colour = "black"),
        panel.grid.major = element_blank(),
        panel.grid.minor = element_blank(),
        panel.border = element_blank(),
        panel.background = element_blank(),
        axis.text.x = element_text(angle = 90, size = 13, face = "bold"),
        axis.text.y = element_text(size = 10,face = "bold")) +
  
        scale_y_continuous(
                          labels = scales::number_format(accuracy = 0.01,
                          decimal.mark = '.'))
        
        print(plot)
}

```
